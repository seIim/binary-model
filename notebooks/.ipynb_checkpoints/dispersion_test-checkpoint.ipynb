```python
def main():
    v_loc   = 0
    v_scale = 1
    key = random.key(11)
    m = jnp.array([0.8]*4000)
    log_g = jnp.array([1]*4000)
    bin_frac = 1.0
    state = generate_orbit_state(key,bin_frac,m,log_g)
    plt.hist(jnp.log10(jnp.abs(state.velocities.v_r[state.binaries])), bins=20, density=True,
            histtype='bar', rwidth=0.980)
    plt.ylabel(r'$\mathrm{f(\log_{10}v_r)}$', fontsize=10)
    plt.xlabel(r'$\log_{10}\mathrm{v_r}$', fontsize=10)
    plt.savefig('../figs/vrdist.pdf', bbox_inches='tight')
    plt.show()
main()
```

```python
def v_disp(v_err, log_g, N_obs, dt_yrs, key):

    dt_day = 365.25 * dt_yrs
    v_loc = 0
    m = jnp.array([0.8]*2000)
    log_g = jnp.array([log_g]*2000)
    bin_frac = 0.5

    state = generate_orbit_state(key,bin_frac,m,log_g)
    masses_masked = state.m_1[state.binaries]
    body_fn = jax.jit(partial(step,dt_days=dt_day))
    
    frac = []
    for i in range(1):
        stacked_v_t = []
        for j in range(N_obs):
            key, _ = random.split(key)
            state = body_fn(key=key,masses_masked=masses_masked,state=state)
            v_t = state.velocities.v_t
            stacked_v_t.append(v_t)
        arr = jnp.array(stacked_v_t)
        arr = jnp.where(arr/v_err > 10, 1, 0)
        arr = jnp.sum(arr, axis=0)
        arr = jnp.where(arr > 0, 1, 0)
        frac = arr.sum()/state.binaries.sum()
        resolved_binaries = arr[ state.binaries == 1 ].sum()
        false_binaries = arr[ state.binaries == 0 ].sum()
        print(resolved_binaries, false_binaries, state.binaries.sum())
        return frac#, false_binares/state.binaries.sum()

Ns =[1, 3, 5, 10, 50, 100]
DTs = [0.5, 1, 2, 4, 8]
frac01 = []
frac05 = []
frac1 = []

key = random.key(11)
for N in Ns:
    for i in range(2):
        key, _ = random.split(key, 2)
        #frac01.append(v_disp(0.1,2.0,N,DTs[2], key))
        frac05.append(v_disp(0.5,2.0,N,DTs[0], key))
        frac1.append(v_disp(1.0,2.0,N,DTs[0], key))
```

```python
frac01, frac05, frac1 = map(jnp.array, [frac01, frac05, frac1])
```

```python
labels = [r'$\sigma_\mathrm{obs} = 1$', r'$\sigma_\mathrm{obs} = 0.5$']
for i,frac in enumerate([frac05, frac1]):
    mu = frac.reshape(len(Ns),5).mean(axis=1)
    unc = frac.reshape(len(Ns),5).std(axis=1)
    plt.errorbar(Ns, mu, yerr=unc, fmt='none', elinewidth=1, ecolor='k')
    plt.scatter(Ns, mu, zorder=10, s=20, edgecolors='k', lw=0.5, label=labels[i])
plt.legend(frameon=False, fontsize=9)
plt.ylabel(r'Detection Fraction ($5\sigma$)', fontsize=11)
plt.xlabel('Number of epochs', fontsize=11)
plt.savefig('../figs/detectfrac.pdf', bbox_inches='tight')
```
